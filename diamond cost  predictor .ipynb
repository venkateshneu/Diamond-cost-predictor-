# Key findings from EDA 

Have decided to use ordinal encoding for the 3 categorical columns cut,
clarity, color as price seem to vary with all these categories 
1. ideal has more occurences indicating ideal cut diamonds were sold more about 92454, highest of all.
The mean price of ideal cut is hte lowest of all cuts. By this we can assume that ideal cut is bought more as the price is low 
2. The descending order in whice clarity vary wrt price is as follows
SI2,SI1,VS2,VS1 ,I1, VVS2,IF,VVS1    
3.  The descending order in whice cUT vary wrt price is as follows 
 Premium, Fair, Very Good, Good, Ideal. Here also Ideal is bought more as its cheaper
4. The descending order in whice color vary wrt price is as follows 
 J, I, H, G, F, E, D

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

```

```python
X.head(2)
```

```python
y.head(2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train
```

```python
# Ordinal encoding transformer
ordinal_encoder = ColumnTransformer(
    transformers=[
        ("cut_encoder", OrdinalEncoder(categories=[['Premium', 'Fair', 'Very Good', 'Good', 'Ideal']]), ["cut"]),
        ("color_encoder", OrdinalEncoder(categories=[['J', 'I', 'H', 'G', 'F', 'E', 'D']]), ["color"]),
        ("clarity_encoder", OrdinalEncoder(categories=[['SI2', 'SI1', 'VS2', 'VS1', 'I1', 'VVS2', 'IF', 'VVS1']]), ["clarity"]),
    ],
    remainder="passthrough"
)

```

```python
from sklearn.linear_model import LinearRegression
# Standard scaling transformer
standard_scaler = ColumnTransformer([("scaling", StandardScaler(), slice(0, 9))], remainder="passthrough")

# Linear regression model
linear_regression_model = LinearRegression()
```

```python
# Define pipeline
pipe = Pipeline([
    ("step1", ordinal_encoder),
    ("step2", standard_scaler),
    ("step3", linear_regression_model)
])
```

```python
pipe
```

```python
pipe.fit(X_train, y_train)
```

```python
predictions = pipe.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
r2_score = r2_score(y_test,predictions )
print("r2_score is :" , r2_score)
```

```python
from sklearn.preprocessing import FunctionTransformer
trf = FunctionTransformer(func = np.log1p)
```

```python
y_train_transformed = trf.fit_transform(y_train)
y_test_transformed = trf.transform(y_test)
```

```python
%%time 
# Pasting 
from sklearn.ensemble import BaggingRegressor
from sklearn.neighbors import KNeighborsRegressor  # Import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline


base_estimator = KNeighborsRegressor()  

bagging_regressor = BaggingRegressor(base_estimator=base_estimator,
                                     n_estimators=10,
                                     max_samples=1.0,  
                                     max_features=1.0, 
                                     bootstrap=False,  
                                     oob_score=False,  
                                     verbose=0,  
                                     random_state=42)

bagging_pipe = Pipeline([
    ("step1", ordinal_encoder),
    ("step2", standard_scaler),
    ("bagging_regressor", bagging_regressor)
])


bagging_pipe.fit(X_train, y_train_transformed)


y_pred_bagging = bagging_pipe.predict(X_test)


r2_bagging = r2_score(y_test_transformed, y_pred_bagging)

print("R-squared for bagging with KNeighborsRegressor:", r2_bagging)
```
